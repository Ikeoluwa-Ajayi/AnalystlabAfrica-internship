## Task 1: Dataset Understanding

Shape: 541,909 rows Ã 8 columns

Data Types:
InvoiceNo: object (categorical)
StockCode: object (categorical)
Description: object (categorical)
Quantity: int64 (numerical)
InvoiceDate: object â converted to datetime 
UnitPrice: float64 (numerical) 
CustomerID: float64 â converted to object
Country: object (categorical)

Numerical features: Quantity, UnitPrice
Categorical features: InvoiceNo, StockCode, Description, CustomerID, Country

Possible primary key: No single column uniquely identifies a row. InvoiceNo + StockCode together form a composite key, since one invoice can contain multiple products.

Dataset Description:
This dataset contains transactional records from a UK-based online retail store, covering December 2010 to December 2011. Each row represents a single product line item within a customer invoice â recording what was purchased, the quantity, unit price, transaction date, and the associated customer and country. Since one invoice can include multiple products, the data is structured at the invoice-line-item level rather than one row per order.

## Task 2: Data Cleaning

### Missing Values

Description: 1,454 missing values â dropped, since there's no logical way to recover a missing product description, and it's a very small fraction (~0.27%) of the data.
CustomerID: 135,080 missing values (~25%) â filled with "Unknown" rather than dropped, since these likely represent guest/unregistered checkouts and dropping them would remove a quarter of the dataset.

### Duplicate Records
Found 5,268 exact duplicate rows (identical across all columns) â likely caused by data entry or export errors rather than real repeat transactions. These were removed using 'drop_duplicates()'.

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

### Data Validation

Negative Quantity (9,725 rows found):
9,251 rows had invoice numbers starting with "C" â legitimate customer cancellations/returns. These were retained, since they represent real business activity, but should be excluded from sales-specific metrics (like revenue or top-selling products) since a return isn't a completed sale.
474 rows did not have a "C" prefix â investigation showed these were internal stock adjustments (descriptions like "damages," "faulty," "check," with unit_price of 0 and no customer attached). These were removed, as they aren't genuine transactions.

Negative UnitPrice (2 rows found):
Both rows were labeled "adjust bad debt" â accounting write-off entries, not real product sales. These were removed.

High UnitPrice outlier (38,970):
Verified as a legitimate, real transaction. Retained.

CustomerID inconsistency:
Initial fill created the literal string "nan" instead of "Unknown" due to converting to string before filling missing values. Fixed using .replace('nan', 'Unknown') to ensure consistent labeling of guest transactions.

## Task 3: Exploratory Data Analysis (EDA)

Explored the following patterns:
1.Top-selling products (by total quantity sold)
2.Highest revenue-generating countries
3.Monthly sales trends
4.Most purchased products (by number of distinct orders)
5.Customer purchasing behavior (total spend and order count per customer)

```python
top_products = df.groupby('description')['quantity'].sum().sort_values(ascending=False).head(10)
top_products
```

```python
highest_revenue = df.groupby('country')['revenue'].sum().sort_values(ascending=False).head(10)
highest_revenue
```

```python
df['invoice_month'] = df['invoice_date'].dt.to_period('M')
monthly_sales = df.groupby('invoice_month')['revenue'].sum()
monthly_sales
```

```python
most_purchased = df.groupby('description')['invoice_no'].nunique().sort_values(ascending=False).head(10)
most_purchased
```

```python
customer_behavior = df.groupby('customer_id').agg(
    total_spent=('revenue', 'sum'),
    total_orders=('invoice_no', 'nunique')
).sort_values('total_spent', ascending=False)

customer_behavior.head(10)
```

```python
customer_behavior_known = customer_behavior[customer_behavior.index != 'Unknown']
customer_behavior_known.head(10)
```

### Key EDA Findings
Revenue is heavily concentrated in the United Kingdom (~Â£8.19M), far ahead of the Netherlands, EIRE, and Germany.
A small number of "hero products" (e.g., the white hanging heart t-light holder) dominate both quantity sold and order frequency.
Monthly revenue trends upward through most of the year; the apparent drop in December 2011 is a data artifact, since the dataset ends mid-month rather than reflecting an actual sales decline.
Order quantities are highly right-skewed â the median order is 3 units, while the mean (9.67) is pulled up by rare large bulk orders and returns.
After excluding "Unknown," a small group of high-value customers account for disproportionately high spend and order counts.

## Task 4: Data Visualization

Five visualizations were created:
Bar chart â Top 10 best-selling products by quantity
Pie chart â Revenue share by top 5 countries
Line chart â Monthly revenue trend
Histogram â Distribution of quantity per transaction
Box plot â Unit price distribution

```python
import matplotlib.pyplot as plt
```

```python
top_products = df.groupby('description')['quantity'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,4))
top_products.plot(kind='bar', color='skyblue')
plt.title('Top 10 Best-Selling Products by Quantity')
plt.xlabel('Product')
plt.ylabel('Total Quantity Sold')
plt.xticks(rotation=75)
plt.show()
```

```python
top_countries = df.groupby('country')['revenue'].sum().sort_values(ascending=False).head(5)

plt.figure(figsize=(7,7))
plt.pie(top_countries, labels=top_countries.index, autopct='%1.1f%%')
plt.title('Revenue Share by Top 5 Countries')
plt.show()
```

```python
monthly_sales = df.groupby('invoice_month')['revenue'].sum()

plt.figure(figsize=(10,6))
monthly_sales.plot(kind='line', marker='o', color='green')
plt.title('Monthly Revenue Trend')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,6))
df['quantity'].plot(kind='hist', bins=50, color='orange', edgecolor='black')
plt.title('Distribution of Quantity per Transaction')
plt.xlabel('Quantity')
plt.ylabel('Frequency')
plt.xlim(-50, 100)  # zoom in, since extreme outliers would squash the chart
plt.show()
```

```python
plt.figure(figsize=(8,6))
df['unit_price'].plot(kind='box')
plt.title('Unit Price Distribution')
plt.ylabel('Unit Price')
plt.ylim(0, 50)  # zoom in on the typical range
plt.show()
```
