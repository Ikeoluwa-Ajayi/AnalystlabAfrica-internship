`date_added` vs `release_year` logic check: Compared whether any title was added to Netflix before its release year (which would be illogical). A small number of rows (mostly TV shows, plus 2 movies) showed a date_added year earlier than release_year. This is likely explained by multi-season shows being dated by their most recent season, and by regional/early release timing for a couple of movies. These rows were retained, as the discrepancy reflects real-world content release patterns rather than a data error.

### Key EDA Findings
- Movies make up roughly 70% of all titles (6,126 Movies vs. 2,664 TV Shows).
- Content additions to Netflix grew sharply between 2015 and 2019 (from under 100 titles/year to over 2,000), before declining slightly in 2020â2021 â likely due to the dataset's collection cutoff rather than an actual slowdown.
- The United States is the top content-producing country, followed by India. "Unknown" also appears prominently, representing missing country data rather than a real production origin.
- TV-MA and TV-14 together account for nearly 70% of all rated content, showing the catalog skews toward mature/general audiences.
- "Dramas, International Movies" is the most common genre combination, reflecting Netflix's global content sourcing strategy.

```python
import matplotlib.pyplot as plt
```

## Task 4: Data Visualization

Five visualizations were created:
1. Bar chart â Movies vs TV Shows distribution
2. Line chart â Number of titles added to Netflix per year
3. Bar chart â Top 5 content-producing countries
4. Pie chart â Distribution of top 6 content ratings
5. Horizontal bar chart â Top 10 most common genre combinations

```python
type_distribution = df['type'].value_counts()

plt.figure(figsize=(6,6))
type_distribution.plot(kind='bar', color=['steelblue', 'salmon'])
plt.title('Movies vs TV Shows on Netflix')
plt.xlabel('Content Type')
plt.ylabel('Number of Titles')
plt.xticks(rotation=0)
plt.show()
```

```python
content_by_year = df['year_added'].value_counts().sort_index()

plt.figure(figsize=(10,6))
content_by_year.plot(kind='line', marker='o', color='green')
plt.title('Number of Titles Added to Netflix per Year')
plt.xlabel('Year')
plt.ylabel('Number of Titles Added')
plt.show()
```

```python
top_countries = df['country'].value_counts().head(5)

plt.figure(figsize=(10,6))
top_countries.plot(kind='bar', color='orange')
plt.title('Top 5 Content-Producing Countries')
plt.xlabel('Country')
plt.ylabel('Number of Titles')
plt.xticks(rotation=75)
plt.show()
```

```python
common_ratings = df['rating'].value_counts().head(6)

plt.figure(figsize=(7,7))
plt.pie(common_ratings, labels=common_ratings.index, autopct='%1.1f%%')
plt.title('Distribution of Top 6 Content Ratings')
plt.show()
```

```python
common_genres = df['listed_in'].value_counts().head(10)

plt.figure(figsize=(10,6))
common_genres.plot(kind='barh', color='purple')
plt.title('Top 10 Most Common Genre Combinations')
plt.xlabel('Number of Titles')
plt.ylabel('Genre Combination')
plt.show()
```
